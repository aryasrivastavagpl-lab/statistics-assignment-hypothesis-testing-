     3.Differentiate between type 1 and type 2 errors

      - Type 1 error : occurs when the null hypothesis is true
      - Type 2 error : occurs when the null hypothesis is false

      .Difference : type 1 is mistakenly claiming a difference exists type 2 is falling to detect a difference that does exists
